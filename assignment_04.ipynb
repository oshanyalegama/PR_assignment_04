```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

# Generate data with make_circles
np.random.seed (5)
X , y = make_circles( n_samples =500 , factor =0.3 , noise =0.1)
```

```python
ax = plt.subplot()
ax.scatter(X[:, 0],X[:, 1], c=y, cmap=plt.cm.Paired)
plt.show()
```

```python
z = np.sum(X**2, axis=1)
ax = plt.subplot(projection='3d')
ax.scatter3D(X[:, 0], X[:, 1], z, c=y, cmap=plt.cm.Paired)
plt.show()
```

```python
X, y = make_circles(n_samples=500, factor=0.5, noise=0.1)
ax = plt.subplot()
ax.scatter(X[:, 0],X[:, 1], c=y, cmap=plt.cm.Paired)
plt.show()
```

```python
X = X**2
z = X[:,0]*X[:,1]
ax = plt.subplot(projection='3d')
ax.scatter3D(X[:, 0], X[:, 1], z, c=y, cmap=plt.cm.Paired)
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

Getting the original data


```python
# Generate data with make_circles
np.random.seed (5)
X , y = make_circles( n_samples =500 , factor =0.3 , noise =0.1)
```

Using normal method

```python
classifier1 = SVC(kernel='linear')
classifier1.fit(X, y)
```

```python
# Generate Test Dataset
X_test, y_test = make_circles(n_samples=100, factor=0.3, noise=0.1)
```

```python
y_pred = classifier1.predict(X_test)
```

```python
print("Test Accuracy is: ", accuracy_score(y_test, y_pred))
```

Using transformation

```python
z = np.sum(X**2, axis=1).reshape(500,1)
X = np.concatenate((X, z), axis=1)
```

```python
classifier2 = SVC(kernel='linear')
classifier2.fit(X,y)
```

```python
X_test, y_test = make_circles(n_samples=100, factor=0.3, noise=0.1)
z_test = np.sum(X_test**2, axis=1).reshape(100,1)
X_test = np.concatenate((X_test, z_test), axis=1)
```

```python
y_pred = classifier2.predict(X_test)
```

```python
print("Test Accuracy is: ", accuracy_score(y, y_pred))
```
